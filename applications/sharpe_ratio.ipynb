### Data :




Note : The rm module wraps custom python functions to compute analysis.

import_data() function returns the price dataframe of all tickers passed as parameters 

Prices source : https://finance.yahoo.com

```python

crypto_data = rm.import_data('2021-01-01','2021-11-27','BTC-USD','ETH-USD','BNB-USD')
```

### Analysis  : 


```python
#compute returns, risk-return performance 
returns = rm.get_return(crypto_data)
returns_tab = rm.return_analysis(returns)
volatility = rm.get_volatility(returns)
Sharpe_Ratio = rm.get_sharpe_ratio(returns_tab,volatility)
```

### Save key results as csv files : Return, Annualized return, Excess Return, Volatility and Annualized volatility.

```python

#Save results as csv files for visualization

returns.to_csv('results/Return_series.csv',index_label="asset")
returns_tab.to_csv('results/Return_Analysis.csv',index_label="asset")
volatility.to_csv('results/Volatility_Analysis.csv',index_label="asset")
Sharpe_Ratio.to_csv('results/Sharpe_Ratio.csv',index_label="asset")
```
